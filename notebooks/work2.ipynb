```python
# extracting from url
import re
import pandas as pd

def extract_location(url):

    if pd.isna(url):
        return pd.Series([None, None, None])

    url = str(url).lower()

    city = None
    state = None
    venue = None

    try:

        parts = url.strip("/").split("/")

        # Example:
        # ['https:', '', 'seatgeek.com',
        # 'oklahoma-city-thunder-tickets',
        # '5-20-2026-oklahoma-city-paycom-center']

        if len(parts) >= 5:

            location_part = parts[4]

            # Remove date
            location_part = re.sub(
                r'^\d{1,2}-\d{1,2}-\d{4}-',
                '',
                location_part
            )

            words = location_part.split("-")

            if len(words) >= 2:

                city = " ".join(words[:2]).title()

                state = " ".join(words[:2]).title()

                venue = " ".join(words[2:]).title()

    except:
        pass

    return pd.Series([city, state, venue])


events_clean[["city","state","venue_name"]] = (
    events_clean["url"].apply(extract_location)
)
```

```python
# ==========================================================
# Create Master Dataset
# ==========================================================

master_df = events_clean.copy()

print("Master Dataset Created")
print(master_df.shape)
```

```python
# ==========================================================
# Extract Performer 1 & Performer 2 IDs
# ==========================================================

import ast
import pandas as pd

def extract_performers(x):

    ids = ast.literal_eval(x)

    performer1 = str(ids[0]) if len(ids) >= 1 else None
    performer2 = str(ids[1]) if len(ids) >= 2 else None

    return pd.Series([performer1, performer2])


master_df[["performer1Id","performer2Id"]] = (
    master_df["performerIds"].apply(extract_performers)
)

master_df[
    ["performerIds","performer1Id","performer2Id"]
].head()
```

```python
# ==========================================================
# Merge Venue Dataset
# ==========================================================

venue_features = [

    "venueId",

    "capacity",
    "capacity_level",

    "venue_score",
    "venue_popularity",
    "venue_popularity_level",

    "addressCountry",
    "addressPostalCode"
]

master_df = master_df.merge(

    venues_clean[venue_features],

    on="venueId",

    how="left"

)

print("Venue Merge Completed")
print(master_df.shape)
```

```python
# ==========================================================
# Merge Performer 1
# ==========================================================

performer_features = [

    "performerId",

    "name",
    "performer_score",
    "performer_popularity",
    "performer_popularity_level",

    "divisionName",
    "divisionShortName",

    "has_home_venue"
]

performer1 = (
    performers_clean[performer_features]
    .add_prefix("performer1_")
)

master_df = master_df.merge(

    performer1,

    left_on="performer1Id",
    right_on="performer1_performerId",

    how="left"

)

print("Performer 1 Merge Completed")
print(master_df.shape)
```

```python
# ==========================================================
# Merge Performer 2
# ==========================================================

performer2 = (
    performers_clean[performer_features]
    .add_prefix("performer2_")
)

master_df = master_df.merge(

    performer2,

    left_on="performer2Id",
    right_on="performer2_performerId",

    how="left"

)

print("Performer 2 Merge Completed")
print(master_df.shape)
```

```python
# ==========================================================
# Merge Summary
# ==========================================================

print("Final Shape :", master_df.shape)

print("\nVenue Matches")
print(master_df["capacity"].notna().sum())

print("\nPerformer 1 Matches")
print(master_df["performer1_name"].notna().sum())

print("\nPerformer 2 Matches")
print(master_df["performer2_name"].notna().sum())
```

```python
master_df.drop(
    columns=[
        "performer1_performerId",               # we are dropping this because these 2 are repeated column made due to merge.
        "performer2_performerId"
    ],
    inplace=True
)

print("Final Shape :", master_df.shape)
```
